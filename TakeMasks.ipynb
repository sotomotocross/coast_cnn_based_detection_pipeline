```python
import cv2
import numpy as np
import json
import os
import matplotlib.pyplot as plt

def loadimages(path = "."):
    return [os.path.join(path, f) for f in sorted(os.listdir(path)) if f.endswith('.png')]
def loadlabels(path = "."):
    return [os.path.join(path, f) for f in sorted(os.listdir(path)) if f.endswith('.json')]

ii=0
ji=0
wi=0
ki=0

path1='FrPNG'
path2='Labels'

filenames=loadimages(path1)
# dir_list_images = sorted(os.listdir(path1))
# print("Files and directories in '", path1, "' :")    
# # print the list 
# print(dir_list_images) 

filelabels=loadlabels(path2)
# dir_list_labels = sorted(os.listdir(path2))
# print("Files and directories in '", path2, "' :")    
# # print the list 
# print(dir_list_labels) 

num=0
for file in filelabels:
    #xx=[]
    
    with open(file) as json_file:
        data = json.load(json_file)
    pnt=data['shapes']
    
    mask=np.zeros((720,1280))
    for j in  range(len(pnt)):
        #print(j)
        pnto=pnt[j]
        xx=[]
        
        pnto=pnto.get('points')
        for i in range(len(pnto)):
            x , y = pnto[i]
            x=int(x)
            y=int(y)
            xx.append((x,y))
        polygon=np.asarray(xx, dtype=None, order=None)
        polygon=np.array([polygon])
    
        #cv2.polylines(mask, polygon, 0, (255,255,255), 3, lineType=cv2.LINE_AA, shift=0)
        cv2.fillPoly(mask, polygon, (255,255,255), lineType=4, shift=0)
    
    #cv2.imshow('result',mask)
    #plt.imshow(mask)
    #plt.show()
    #cv2.waitKey(0)
    #cv2.destroyAllWindows()
    cv2.imwrite('Masks/Mask'+str(ki)+str(wi)+str(ji)+str(ii)+'.png',mask)
    #cv2.imwrite(str(0)+str(ki)+str(wi)+str(ji)+str(ii)+'.png',mask)
    ii=ii+1
    num=num+1
    if ii==10:
        ii=0
        ji=ji+1
    if ji==10:
        ji=0
        wi=wi+1
    if wi==10:
        wi=0
        ki=ki+1
```
