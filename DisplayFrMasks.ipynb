```python
import cv2
import numpy as np
#import json
import os
def loadimages(path = "."):
    return [os.path.join(path, f) for f in sorted(os.listdir(path)) if f.endswith('.png')]

i=0
j=0
w=0
k=0

# width and height can be modified according to the resolution of the frame .png files
dim=(224,224)
num=0
path1='ResizedFr'
path2='ResizedMasks'
filenames=loadimages(path1)
filelabels=loadimages(path2)

def create_blank(width, height, rgb_color=(0, 0, 0)):
    """Create new image(numpy array) filled with certain color in RGB"""
    # Create black blank image
    image = np.zeros((height, width, 3), np.uint8)

    # Since OpenCV uses BGR, convert the color first
    color = tuple(reversed(rgb_color))
    # Fill image with color
    image[:] = color

    return image

red = (255, 0, 0)

for file in filenames:
    image=cv2.imread(filenames[num])
    copy_image=np.copy(image)
    
    masked=cv2.imread(filelabels[num])
    masked_rsz=cv2.resize(masked, dim, interpolation = cv2.INTER_AREA)
    copy_mask=np.copy(masked_rsz)
    
    ww = copy_mask.shape[1]
    hh = copy_mask.shape[0]
    red_mask = create_blank(ww, hh, rgb_color=red)
    copy_mask=cv2.bitwise_and(copy_mask,red_mask)
    
    combo_image=cv2.addWeighted(copy_image, 1, copy_mask,0.35 ,1)
    cv2.imshow('result'+str(num),combo_image)
    cv2.waitKey(100)
    cv2.destroyAllWindows()
    num=num+1
    #if num>15:
        #break

    filename = str(num)+str(k)+str(w)+str(j)+str(i)
    cv2.imwrite('./Display_before_classification/'+str(filename)+'.png', combo_image)
    i=i+1
    if i==10:
        i=0
        j=j+1
    if j==10:
        j=0
        w=w+1
    if w==10:
        w=0
        k=k+1
    if k==10:
        k=0
        num=num+1
```
